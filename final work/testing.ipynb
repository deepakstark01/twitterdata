```python

import pytz

def timestamp_converter(timestamp):
    timestamp = str(timestamp)
    converted_stamp = timestamp[:5] + timestamp[8:]
    return datetime.fromtimestamp(int(converted_stamp))


def stamp_to_timezone(timestamp, timezone):
    dt = timestamp_converter(timestamp)
    dt = dt.replace(tzinfo=pytz.utc)
    new_dt = dt.astimezone(pytz.timezone(timezone))
    return new_dt.strftime('%Y-%m-%d %I:%M:%p')


print(stamp_to_timezone('1660843944407', 'Europe/Rome'))
```
